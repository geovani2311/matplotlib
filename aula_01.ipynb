```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x ** 2 
```

```python
y
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## simples

```python
plt.plot(x,y)
```

## customizando

```python
plt.plot(x, y, "red")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title("Titulo")
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, "r--")
plt.subplot(1,2,2)
plt.plot(y, x, "g*--")
```

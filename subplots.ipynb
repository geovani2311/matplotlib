## subplots

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2 
```

```python
(1,2) # = tupla quando entre parenteses
```

```python
a,b = (1,2)
```

```python
fig, ax = plt.subplots()

ax.plot(x, y, "r--")
ax.set_title("Titulo do nosso subplot")
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2)

ax[0].plot(x,y,"red")
ax[1].plot(x,y,"green")
```

## tight_layout

```python
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(30,8))

for axis in ax:
    axis.plot(x,y,"r")

plt.tight_layout()
```

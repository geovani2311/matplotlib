## Orientação a objetos no Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2 
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])  #esquerda, inferior, largura e altura

axes.plot(x, y, "blue")
axes.set_xlabel("Label x")
axes.set_ylabel("Label y")
axes.set_title("Titulo")
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])  #esquerda, inferior, largura e altura
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes.plot(x, y, "blue")
axes.set_xlabel("Label x")
axes.set_ylabel("Label y")
axes.set_title("Titulo")
                                       
axes2.plot(y, x, "red")
axes2.set_xlabel("Label x2")
axes2.set_ylabel("Label y2")
axes2.set_title("Titulo2")
```
